```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
file='111.npy'
```

```python
#загрузка исходного файла
np_arr = np.load(file)
```

```python
I0=np_arr[0] #зможно использовать первый кадр без нагрузки для коррекции 
I=np_arr[10]
plt.figure(figsize=(10,10)) # размер изображения
img_grey = plt.imshow(I,cmap='gray',interpolation='bilinear')
plt.savefig('foo_gray.png')

#imgplot_rgb = plt.imshow(I, cmap='gray')
#interpolation Supported values are 'none', 'antialiased', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36', 'hanning', 'hamming',
#'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos', 'blackman'.
```

```python
colormin = np.array((12), np.uint8) # нижняя граница
colormax = np.array((220), np.uint8) # верхняя граница
img = cv2.imread('foo_gray.png')
img_gray = cv2.cvtColor( img, cv2.COLOR_BGR2GRAY ) 
#thresh = cv2.inRange( img_gray, colormin, colormax )
thresh = cv2.inRange( img, colormin, colormax )

 # отображаем контуры поверх изображения
contours, hierarchy = cv2.findContours( thresh.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#index_cont отображение определенного замкнутого контура, например 5
index_cont=4
II=cv2.drawContours( img, contours, index_cont, (240), 2, cv2.LINE_AA, hierarchy, 1)
plt.figure(figsize=(10,10)) # размер изображения
plt.imshow(II)
```
